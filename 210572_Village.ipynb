```python
import numpy as np
import cv2 as cv

from google.colab.patches import cv2_imshow
```

```python
# Reading the image

img_num = 1
img = cv.imread(f'./cctv{img_num}.JPG')
```

```python
cv2_imshow(img)
```

```python
width = img.shape[1]
height = img.shape[0]
```

```python
gray = cv.cvtColor(img, code = cv.COLOR_BGR2GRAY)
```

```python
if(img_num == 1):
  final = np.array(gray, copy = True)

  # Changing the contrast of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = gray[i][j] * 2

  # Changing the brightness of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = (final[i][j] / 255)**(1/2) * 255

  for i in range(0, height):
    for j in range(0, width):
      if(final[i][j] < 0):
        final[i][j] = 0;
      if(final[i][j] > 255):
        final[i][j] = 255

  cv2_imshow(final)
```

```python
if(img_num == 2):
  final = np.array(gray, copy = True)

  # Changing the contrast of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = gray[i][j] // 3

  # Changing the brightness of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = (final[i][j] / 255)**(1/3) * 255


  for i in range(0, height):
    for j in range(0, width):
      if(final[i][j] < 0):
        final[i][j] = 0
      if(final[i][j] > 255):
        final[i][j] = 255

  cv2_imshow(final)
```

```python
if(img_num == 3):
  final = np.array(gray, copy = True)

  # Changing the contrast of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = gray[i][j] * 2

  # Changing the brightness of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = (final[i][j] / 255)**(1/2) * 255

  for i in range(0, height):
    for j in range(0, width):
      if(final[i][j] < 0):
        final[i][j] = 0;
      if(final[i][j] > 255):
        final[i][j] = 255

  cv2_imshow(final)
```

```python
if(img_num == 4):
  final = np.array(gray, copy = True)

  # # Changing the contrast of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = gray[i][j] * 1.5

  # Changing the brightness of the image
  for i in range(0, height):
    for j in range(0, width):
      final[i][j] = (final[i][j] / 255)**(1/2) * 255

  for i in range(0, height):
    for j in range(0, width):
      if(final[i][j] < 0):
        final[i][j] = 0;
      if(final[i][j] > 255):
        final[i][j] = 255

  cv2_imshow(final)
```

```python
cv.imwrite(f'./enhanced-cctv{img_num}.jpg', final)
```
